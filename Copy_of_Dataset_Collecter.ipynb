**Auto Download Correct ChromeDriver**

```python
import re, json, requests, os, subprocess

chrome_version = subprocess.check_output(["google-chrome", "--version"]).decode()
print("Google Chrome Version:", chrome_version)

major_version = re.search(r"(\d+)\.", chrome_version).group(1)
print("Chrome Major Version:", major_version)

data = requests.get(
    "https://googlechromelabs.github.io/chrome-for-testing/known-good-versions-with-downloads.json"
).json()

driver_url = None
for v in data["versions"]:
    if v["version"].startswith(major_version + "."):
        for d in v["downloads"]["chromedriver"]:
            if d["platform"] == "linux64":
                driver_url = d["url"]
                break
    if driver_url:
        break

print("Downloading Driver From:", driver_url)

!wget -O /tmp/chromedriver.zip $driver_url
!unzip -o /tmp/chromedriver.zip -d /tmp/
!mv /tmp/chromedriver-linux64/chromedriver /usr/local/bin/chromedriver
!chmod +x /usr/local/bin/chromedriver

print("Chrome + Driver Ready!")
```

```python
import requests

print("Downloading phishing URLs...")

data = requests.get("https://openphish.com/feed.txt").text.strip().split("\n")

with open("phishing_urls.txt", "w") as f:
    for url in data:
        f.write(url + "\n")

print("phishing_urls.txt created")
print("Total URLs saved:", len(data))
```

```python
import random

keywords = [
    "paytm","phonepe","upi",
    "sbi","yono","onlinesbi",
    "hdfc","icici","axis","kotak",
    "irctc","lic","uidai","rbi",
    "flipkart","amazon","myntra",
    "login","secure","verify"
]

indian = []
others = []

with open("phishing_urls.txt", "r") as f:
    urls = f.read().splitlines()

print("Total URLs loaded:", len(urls))

for u in urls:
    ul = u.lower()
    if any(k in ul for k in keywords):
        indian.append(u)
    else:
        others.append(u)

print("Indian targeted phishing found:", len(indian))

indian = list(set(indian))
others = list(set(others))

random.shuffle(indian)
random.shuffle(others)

final = []
final.extend(indian[:2500])
final.extend(others[:1500])
random.shuffle(final)

with open("phishing_urls_final.txt", "w") as f:
    for x in final:
        f.write(x + "\n")

print("Saved phishing_urls_final.txt")
print("Final selected phishing count:", len(final))
```

**Screenshot Automation**

```python
import os, time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

CHROME_PATH = "/usr/bin/google-chrome"
CHROMEDRIVER_PATH = "/usr/local/bin/chromedriver"

def driver_setup():
    options = webdriver.ChromeOptions()
    options.add_argument("--headless=new")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--disable-gpu")
    options.add_argument("--ignore-certificate-errors")
    options.add_argument("--window-size=1366,768")
    options.add_argument("--lang=en-IN")
    options.add_argument("user-agent=Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome Safari")
    options.binary_location = CHROME_PATH
    return webdriver.Chrome(service=Service(CHROMEDRIVER_PATH), options=options)

def load(filename):
    return open(filename).read().splitlines()

def capture(urls, folder):
    os.makedirs(folder, exist_ok=True)
    d = driver_setup()
    i = 1
    for u in urls:
        try:
            d.get(u)
            time.sleep(4)
            path = f"{folder}/{folder.split('/')[-1]}_{i}.png"
            d.save_screenshot(path)
            print("Saved:", path)
            i += 1
        except Exception as e:
            print("Failed:", u, "| Reason:", e)
    d.quit()

phish = load("phishing_urls.txt")
legit = load("legit_urls.txt")

capture(phish, "dataset/phishing")
capture(legit, "dataset/legit")
```
